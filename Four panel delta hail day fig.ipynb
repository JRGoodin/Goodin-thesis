# Notebook for creating four panel delta hail day figures using png images

### Import necessary libaries (matplotlib for plotting, PIL Image for reading in and showing images)

```python
import matplotlib.pyplot as plt
from PIL import Image
```

# Create 2x2 panel for only severe [â‰¥ 2.5 cm] HAIL_MAX2D

## Read in .png images from scratch directory and assign to variables

```python
delta_mid4p5 = Image.open('/home/scratch/jgoodin/sev_hail_day_climos/delta_HIST_mid4p5_sev_hail_days_panel_fig.png')
delta_mid8p5 = Image.open('/home/scratch/jgoodin/sev_hail_day_climos/delta_HIST_mid8p5_sev_hail_days_panel_fig.png')
delta_end4p5 = Image.open('/home/scratch/jgoodin/sev_hail_day_climos/delta_HIST_end4p5_sev_hail_days_panel_fig.png')
delta_end8p5 = Image.open('/home/scratch/jgoodin/sev_hail_day_climos/delta_HIST_end8p5_sev_hail_days_panel_fig.png')
```

## Use imshow to open images and create 2x2 subplot, assign each image to place 1-4

```python
fig = plt.figure(figsize = (14, 10)) #Create a 14x10 figure
plt.subplot(2, 2, 1) #Assign image to position 1 of 2x2 subplot
plt.imshow(delta_mid4p5) #Display image (contained in variable listed above)
plt.axis('off') #Remove axis automatically included by maptlotlib- we just want the image
ann = plt.annotate('a)', xy = (30., 50.), fontsize = 30, bbox=dict(boxstyle="round,pad=0.3", fc = 'lightblue')) #Annotate first panel as "a"

plt.subplot(2, 2, 2) #Assign image to position 2 of 2x2 subplot
plt.imshow(delta_mid8p5) #Display image (contained in variable listed above)
plt.axis('off') #Remove axis automatically included by maptlotlib
ann1 = plt.annotate('b)', xy = (30., 50.), fontsize = 30, bbox=dict(boxstyle="round,pad=0.3", fc = 'lightblue')) #Annotate second panel as "b"

plt.subplot(2, 2, 3) #Assign image to position 3 of 2x2 subplot
plt.imshow(delta_end4p5) #Display image (contained in variable listed above)
plt.axis('off') #Remove axis automatically included by maptlotlib
ann2 = plt.annotate('c)', xy = (30., 50.), fontsize = 30, bbox=dict(boxstyle="round,pad=0.3", fc = 'lightblue')) #Annotate third panel as "c"

plt.subplot(2, 2, 4) #Assign image to position 4 of 2x2 subplot
plt.imshow(delta_end8p5) #Display image (contained in variable listed above)
plt.axis('off') #Remove axis automatically included by maptlotlib
ann3 = plt.annotate('d)', xy = (30., 50.), fontsize = 30, bbox=dict(boxstyle="round,pad=0.3", fc = 'lightblue')) #Annotate fourth panel as "d"

fig.patch.set_edgecolor('black') #Outline figure with black line
fig.patch.set_linewidth('3.5') #Set linewidth of outline
plt.tight_layout() #Automatically spaces panels out to match figure size

plt.savefig('/home/scratch/jgoodin/sev_hail_day_climos/four_panel_delta_sev.png') #Save figure as .png to scratch directory
```
