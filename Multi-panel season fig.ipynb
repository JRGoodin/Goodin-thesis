```python
import matplotlib.pyplot as plt
from PIL import Image
```

## Read in min, mean, max, and stdev .png images from scratch directory and assign to variables

```python
DJF_min = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/mins/DJF_min.png')
DJF_mean = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/means/DJF_mean.png')
DJF_max = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/max/DJF_max.png')
DJF_stdev = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/stdev/DJF_stdev.png')
MAM_min = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/mins/MAM_min.png')
MAM_mean = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/means/MAM_mean.png')
MAM_max = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/max/MAM_max.png')
MAM_stdev = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/stdev/MAM_stdev.png')
JJA_min = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/mins/JJA_min.png')
JJA_mean = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/means/JJA_mean.png')
JJA_max = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/max/JJA_max.png')
JJA_stdev = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/stdev/JJA_stdev.png')
SON_min = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/mins/SON_min.png')
SON_mean = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/means/SON_mean.png')
SON_max = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/max/SON_max.png')
SON_stdev = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/stdev/SON_stdev.png')
```
