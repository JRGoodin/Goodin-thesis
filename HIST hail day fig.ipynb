# Notebook for creating four panel delta hail day figures using png images

### Import necessary libaries (matplotlib for plotting, PIL Image for reading in and showing images)

```python
import matplotlib.pyplot as plt
from PIL import Image
```

## Create 1x2 panel for HIST sev hail days

### Read in .png images from scratch directory and assign to variables

```python
hist_sev = Image.open('/home/scratch/jgoodin/sev_hail_day_climos/HIST_annual_sev_hail_days_inches.png')
hist_4cm = Image.open('/home/scratch/jgoodin/sev_hail_day_climos/HIST_annual_4cm_hail_days_inches.png')
```

## Use imshow to open images and create 1x2 subplot, assign each image to place 1 or 2

```python
fig = plt.figure(figsize = (20, 12)) #Create a 12x8 figure
plt.subplot(1, 2, 1) #Assign image to position 1 of 1x2 subplot
plt.imshow(hist_sev) #Display image (contained in variable listed above)
plt.axis('off') #Remove axis automatically included by maptlotlib- we just want the image
ann = plt.annotate('a)', xy = (20., 50.), fontsize = 30, bbox=dict(boxstyle="round,pad=0.3", fc = 'lightblue')) #Annotate first panel as "a"

plt.subplot(1, 2, 2) #Assign image to position 2 of 1x2 subplot
plt.imshow(hist_4cm) #Display image (contained in variable listed above)
plt.axis('off') #Remove axis automatically included by maptlotlib
ann1 = plt.annotate('b)', xy = (20., 50.), fontsize = 30, bbox=dict(boxstyle="round,pad=0.3", fc = 'lightblue')) #Annotate second panel as "b"

plt.subplots_adjust(wspace=0, hspace=0)
fig.patch.set_edgecolor('black') #Outline figure with black line
fig.patch.set_linewidth('3.5') #Set linewidth of outline
#plt.tight_layout() #Automatically spaces panels out to match figure size
```
