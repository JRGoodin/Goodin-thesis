## Find deltas between HIST and FUTR for each season (FUTR - HIST)

### DJF deltas

```python
delta_DJF_mid4p5 = mid4p5_DJF.HAIL_MAX2D - hist_DJF.HAIL_MAX2D
delta_DJF_mid8p5 = mid8p5_DJF.HAIL_MAX2D - hist_DJF.HAIL_MAX2D #Find DJF deltas for each climate epoch
delta_DJF_end4p5 = end4p5_DJF.HAIL_MAX2D - hist_DJF.HAIL_MAX2D
delta_DJF_end8p5 = end8p5_DJF.HAIL_MAX2D - hist_DJF.HAIL_MAX2D
```

### MAM deltas

```python
delta_MAM_mid4p5 = mid4p5_MAM.HAIL_MAX2D - hist_MAM.HAIL_MAX2D
delta_MAM_mid8p5 = mid8p5_MAM.HAIL_MAX2D - hist_MAM.HAIL_MAX2D #Find MAM deltas for each climate epoch
delta_MAM_end4p5 = end4p5_MAM.HAIL_MAX2D - hist_MAM.HAIL_MAX2D
delta_MAM_end8p5 = end8p5_MAM.HAIL_MAX2D - hist_MAM.HAIL_MAX2D
```

### JJA deltas

```python
delta_JJA_mid4p5 = mid4p5_JJA.HAIL_MAX2D - hist_JJA.HAIL_MAX2D
delta_JJA_mid8p5 = mid8p5_JJA.HAIL_MAX2D - hist_JJA.HAIL_MAX2D #Find JJA deltas for each climate epoch
delta_JJA_end4p5 = end4p5_JJA.HAIL_MAX2D - hist_JJA.HAIL_MAX2D
delta_JJA_end8p5 = end8p5_JJA.HAIL_MAX2D - hist_JJA.HAIL_MAX2D
```

### SON deltas

```python
delta_SON_mid4p5 = mid4p5_SON.HAIL_MAX2D - hist_SON.HAIL_MAX2D
delta_SON_mid8p5 = mid8p5_SON.HAIL_MAX2D - hist_SON.HAIL_MAX2D #Find JJA deltas for each climate epoch
delta_SON_end4p5 = end4p5_SON.HAIL_MAX2D - hist_SON.HAIL_MAX2D
delta_SON_end8p5 = end8p5_SON.HAIL_MAX2D - hist_SON.HAIL_MAX2D
```

### Import netcdf file containing lat/lon coords for WRF files (use this for plotting)

```python
coords = xr.open_dataset('lat_lon.nc') #Open .nc file containing WRF lat/lon coords
```

### Assign lat/lon coords from file to variables

```python
lats = coords.CLAT.values[0, :, :]
lons = coords.CLONG.values[0, :, :]
```

## Plot each delta on map of CONUS

```python
fig = plt.figure(figsize= (14,10)) #Plot a figure
gs1 = gridspec.GridSpec(2, 1, height_ratios=[1, .05], bottom=.05, top=.95, wspace=.1)
map_proj = ccrs.LambertConformal(central_longitude = -100, central_latitude = 35) #Set projection, central lat/lon
ax = plt.subplot(projection = map_proj)
#ax.set_title('Maximum diameter hail for 2005', fontsize = 22)
ax.add_feature(cfeature.LAND.with_scale('10m')) #Add land
countries_shp = shpreader.natural_earth(resolution='50m',
                                     category='cultural',
                                     name='admin_0_countries')
for country, info in zip(shpreader.Reader(countries_shp).geometries(), 
                             shpreader.Reader(countries_shp).records()):
        if info.attributes['NAME_LONG'] != 'United States':

            ax.add_geometries([country], ccrs.PlateCarree(),
                             facecolor='lightgrey', edgecolor='k', zorder=6)
ax.add_feature(cfeature.NaturalEarthFeature('physical', 'coastline', '50m', edgecolor='k', 
                                                facecolor='None'), zorder=8) 
ax.add_feature(cfeature.BORDERS.with_scale('10m')) #Add state borders
ax.add_feature(cfeature.STATES.with_scale('10m'), facecolor = 'none') #Add US states
ax.add_feature(cfeature.NaturalEarthFeature('physical', 'ocean', '50m', edgecolor='face', 
                                                facecolor='lightsteelblue'), zorder=6)
lakes_shp = shpreader.natural_earth(resolution='50m',
                                     category='physical',
                                     name='lakes')
for lake, info in zip(shpreader.Reader(lakes_shp).geometries(), 
                             shpreader.Reader(lakes_shp).records()):
        name = info.attributes['name']
        if name == 'Lake Superior' or name == 'Lake Michigan' or \
           name == 'Lake Huron' or name == 'Lake Erie' or name == 'Lake Ontario':
            
            ax.add_geometries([lake], ccrs.PlateCarree(),
                             facecolor='lightsteelblue', edgecolor='k', zorder=6)
ax.set_extent([240,287,22,50]) #Set lat/lon boundaries
lrFig = ax.pcolormesh(lons, lats, delta_JJA_end4p5, vmin = -6, vmax = 6, cmap = 'seismic', transform = ccrs.PlateCarree())
#ax.text(-126, 32.5, 'DJF', fontsize = 65, va='bottom', ha='center', rotation='vertical', transform=ccrs.PlateCarree())
plt.title(" JJA Days  1 in (END 4.5 - HIST)", fontsize = 42)
ax1 = plt.subplot(gs1[1, 0])
cb = plt.colorbar(lrFig, cax = ax1, ticks = [-6, -3, 0, 3, 6], orientation = 'horizontal', extend = 'both')
#cb.set_label('  Days', fontsize = 30)
cb.ax.tick_params(labelsize = 32)
plt.subplots_adjust(bottom=0.03)
#plt.tight_layout()
plt.savefig('/home/scratch/jgoodin/sev_hail_day_climos/seasonal_deltas/delta_HIST_end4p5_JJA_hail_days.png') #**Make sure to save results to png!**
```
