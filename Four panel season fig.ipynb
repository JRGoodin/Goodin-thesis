# Notebook for creating a four panel figure using png images

### Import necessary libaries (matplotlib for plotting, PIL Image for reading in and showing images)

```python
import matplotlib.pyplot as plt
from PIL import Image
```

## Read in .png images from scratch directory and assign to variables

```python
DJF = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/max/DJF_max_title.png') #Read in DJF max .png file from seasonal mesh directory
MAM = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/max/MAM_max_title.png') #Read in MAM max .png file from seasonal mesh directory
JJA = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/max/JJA_max_title.png') #Read in JJA max .png file from seasonal mesh directory
SON = Image.open('/home/scratch/jgoodin/seasonal_mesh_climos/max/SON_max_title.png') #Read in SON max .png file from seasonal mesh directory
```

## Use imshow to open images and create 2x2 subplot, assign each image to place 1-4

```python
fig = plt.figure(figsize = (14, 10)) #Create a 14x10 figure
plt.subplot(2, 2, 1) #Assign image to position 1 of 2x2 subplot
plt.imshow(DJF) #Display image (contained in variable listed above)
plt.axis('off') #Remove axis automatically included by maptlotlib- we just want the image
ann = plt.annotate('a)', xy = (75., 100.), fontsize = 30, bbox=dict(boxstyle="round,pad=0.3", fc = 'lightblue')) #Annotate first panel as "a"

plt.subplot(2, 2, 2) #Assign image to position 2 of 2x2 subplot
plt.imshow(MAM) #Display image (contained in variable listed above)
plt.axis('off') #Remove axis automatically included by maptlotlib
ann1 = plt.annotate('b)', xy = (75., 100.), fontsize = 30, bbox=dict(boxstyle="round,pad=0.3", fc = 'lightblue')) #Annotate second panel as "b"

plt.subplot(2, 2, 3) #Assign image to position 3 of 2x2 subplot
plt.imshow(JJA) #Display image (contained in variable listed above)
plt.axis('off') #Remove axis automatically included by maptlotlib
ann2 = plt.annotate('c)', xy = (75., 100.), fontsize = 30, bbox=dict(boxstyle="round,pad=0.3", fc = 'lightblue')) #Annotate third panel as "c"

plt.subplot(2, 2, 4) #Assign image to position 4 of 2x2 subplot
plt.imshow(SON) #Display image (contained in variable listed above)
plt.axis('off') #Remove axis automatically included by maptlotlib
ann3 = plt.annotate('d)', xy = (75., 100.), fontsize = 30, bbox=dict(boxstyle="round,pad=0.3", fc = 'lightblue')) #Annotate fourth panel as "d"

fig.patch.set_edgecolor('black') #Outline figure with black line
fig.patch.set_linewidth('3.5') #Set linewidth of outline
plt.tight_layout() #Automatically spaces panels out to match figure size

plt.savefig('/home/scratch/jgoodin/seasonal_mesh_climos/max/four_panel_seasonal_max.png') #Save figure as .png to scratch directory
```
